```python
import cv2
from google.colab import files

# Upload your image to Google Colab
uploaded = files.upload()

# Choose the filename of the uploaded image
filename = list(uploaded.keys())[0]

# Read the uploaded image
image = cv2.imread(filename)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Invert the grayscale image
inverted_image = cv2.bitwise_not(gray_image)

# Apply a bilateral filter for a smoother, clearer effect
smooth_image = cv2.bilateralFilter(inverted_image, d=9, sigmaColor=75, sigmaSpace=75)

# Invert the smoothed image
inverted_smooth_image = cv2.bitwise_not(smooth_image)

# Create the pencil sketch by blending the original color image with the inverted smoothed image
sketch = cv2.divide(255 - inverted_smooth_image, 255, scale=256)

# Save the pencil sketch
sketch_filename = 'pencil_sketch.jpg'
cv2.imwrite(sketch_filename, sketch)

# Display the original image and the pencil sketch
from IPython.display import Image, display
display(Image(filename))
display(Image(sketch_filename))
```
